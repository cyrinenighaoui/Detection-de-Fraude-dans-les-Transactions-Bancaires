```python
#lecture fichier csv
import pandas as pd
df = pd.read_csv('creditcard_2023.csv')
df
```

```python
df["Class"].value_counts()
```

```python
df.info()  
```

```python
import matplotlib.pyplot as plt
df['Class'].value_counts().plot(kind='bar')
```

```python
#graphique boxplot class et montant avec matplotlib
plt.figure(figsize=(8,5))
plt.boxplot([df[df['Class'] == 0]['Amount'], df[df['Class'] == 1]['Amount']], labels=['Normal', 'Fraude'])
plt.title('Boxplot du montant des transactions par classe')   
```

```python
import seaborn as sns
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("CorrÃ©lations entre les variables")
plt.show()
plt.show()
```
